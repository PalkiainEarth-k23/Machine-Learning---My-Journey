```python
import pandas as pd
from sklearn.datasets import load_iris
Flower = load_iris()
dir(Flower)
```

```python
df = pd.DataFrame(Flower.data, Flower.target)
df.head()
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
model_list = {
    'RF': {
        'model': RandomForestClassifier(),
        'parameters' : {'n_estimators' : [1,3,6,10]}
},
    'LR': {
            'model' : LogisticRegression(),
            'parameters' : {'l1_ratio' : [0.2, 0.5, 0.8],
                            '' : [0.01, 0.1, 1.0, 10.0]}
},
    'GNB' : {
            'model' : GaussianNB(),
            'parameters' :{}
},
    'MNB': {
           'model' : MultinomialNB(),
            'parameters' : {'alpha': [0.0, 0.01, 0.1, 0.5, 1.0, 2.0],
        'fit_prior': [True, False] }
},
    'DT' : {
           'model' : DecisionTreeClassifier(),
            'parameters' :{'criterion': ['gini', 'entropy'],
        'max_depth': [None, 3, 5, 10, 15],}
}
}
```

```python
score = []
from sklearn.model_selection import GridSearchCV
for model_name, model_info in model_list.items():
    clf = GridSearchCV(
        estimator=model_info['model'], 
        param_grid=model_info['parameters'], 
        cv=5, 
        scoring='accuracy',
        n_jobs=-1
    )
    clf.fit(Flower.data, Flower.target)
    score.append({
        'model': model_name,
        'Best Parameters': clf.best_params_,
        'Best Validation Accuracy': clf.best_score_
    })

df2 = pd.DataFrame(score)
df2.head()
```
