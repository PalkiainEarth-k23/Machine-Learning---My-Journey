```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/srini/Downloads/canada_per_capita_income.csv")
x = df['year'].to_numpy() - 1970
y = df['per_capita_income_$'].to_numpy()

m = 0.0
b = 0.0

learning_rate = 0.0002
iterations = 50000

n = len(x)
previous_cost = float('inf')

for i in range(iterations):
    y_predicted = m*x + b
    cost = 1/n * np.sum((y - y_predicted) ** 2)
    md = -2/n * np.sum(x*(y - y_predicted))
    bd = -2/n * np.sum((y - y_predicted))

    if abs(previous_cost - cost) < 0.01:
        break
    previous_cost = cost
    
    m = m - learning_rate*md
    b = b - learning_rate*bd

print(f"The error(cost) has reached its minimum value {cost} at {i}th iteration and equation of the line is y = {m:.2f}*x + {b:.2f}")
```
