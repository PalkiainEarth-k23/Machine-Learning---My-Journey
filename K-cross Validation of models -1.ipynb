```python
import pandas as pd
import sklearn
from sklearn.datasets import load_iris
Flowers= load_iris()
```

```python
dir(Flowers)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
Results_L = cross_val_score(LogisticRegression(max_iter=1000), Flowers.data, Flowers.target, cv=6)
Results_L
```

```python
Results_RF = cross_val_score(RandomForestClassifier(), Flowers.data, Flowers.target, cv=6)
Results_RF
```

```python
Results_SVC = cross_val_score(SVC(), Flowers.data, Flowers.target, cv=6)
Results_SVC 
```

```python
import numpy as np
```

```python
Average_Score_L = np.mean(Results_L)
Average_Score_RF = np.mean(Results_RF)
Average_Score_SVC  = np.mean(Results_SVC)
```

```python
print(Average_Score_L, Average_Score_RF ,Average_Score_SVC 
```
