```python
import sklearn 
from sklearn.cluster import KMeans
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
```

```python
Flower = load_iris()
```

```python
dir(Flower)
```

```python
df = pd.DataFrame(Flower.data,columns=Flower.feature_names)
df.head()
```

```python
Final = df.drop(['sepal length (cm)', 'sepal width (cm)'], axis = 'columns')
Final.head()
```

```python
scaler_length = MinMaxScaler()
scaler_length.fit(Final[['petal length (cm)']])
Final['Petal_Length'] = scaler_length.transform(Final[['petal length (cm)']])
scaler_width = MinMaxScaler()
scaler_width.fit(Final[['petal width (cm)']])
Final['Petal_Width'] = scaler_width.transform(Final[['petal width (cm)']])
Final.head()
```

```python
Necessary = Final.drop(['petal length (cm)', 'petal width (cm)'], axis = 'columns')
Necessary.head()
```

```python
KM = KMeans(n_clusters = 3)
y_predict = KM.fit_predict(Necessary[['Petal_Length', 'Petal_Width']])
```

```python
Necessary['Cluster'] = y_predict
Necessary.head()
```
