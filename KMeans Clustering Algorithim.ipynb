```python
import sklearn 
from sklearn.cluster import KMeans
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("C:/Users/srini/Downloads/income.csv")
df.head()
```

```python
Final = df.drop(['Name'], axis='columns')
Final.head()
```

```python
scaler = MinMaxScaler()
scaler = MinMaxScaler()
scaler.fit(Final[['Age', 'Income($)']])
Final[['Age', 'Income($)']] = scaler.transform(Final[['Age', 'Income($)']])
Final.head()
```

```python
KM = KMeans(n_clusters = 3)
y_predict = KM.fit_predict(Final)
y_predict
```

```python
Final['cluster'] = y_predict
Final.head()
```

```python
df1 = Final[Final.cluster==0]
df2 = Final[Final.cluster==1]
df3 = Final[Final.cluster==2]
plt.scatter(df1['Age'],df1['Income($)'],color='green')
plt.scatter(df2['Age'],df2['Income($)'],color='red')
plt.scatter(df3['Age'],df3['Income($)'],color='black')
plt.xlabel('Age')
plt.ylabel('Income($)')
```
